```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar el archivo 
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

```python
#Filtro por objeto
filtro1 = df[(df["NOMBRE_VENDEDOR"]) == "LETICIA RAMIREZ HERNANDEZ"]
filtro1.to_csv('filtro1.csv')
filtro1
```

```python
#Filtro por filas 
filtro2 = df.iloc[:3]
filtro2.to_csv('filtro2.csv')
filtro2
```

```python
#Filtro por columnas
filtro3 = df.iloc[:, 0:5]
filtro3.to_csv('filtro3.csv')
filtro3
```

```python
#Cargar el archivo
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=1)
df1
```

```python
#Filtro de filas vs columnas 
filtro4 = df1.loc[["2022-01-05", "2022-12-23"], ["NOMBRE_VENDEDOR", "MARGEN_TOTAL_CALCULADO"]]
filtro4.to_csv('filtro4.csv')
filtro4

```

```python
#Filtro por cabecera 
filtro5 = df.head()
filtro5.to_csv('filtro5.csv')
filtro5
```

```python
#Filtrado por cola
filtro6 = df.tail()
filtro6.to_csv('filtro6.csv')
filtro6
```

```python
#Filtro por comparacion 
filtro7 = df[df["MARGEN_TOTAL_CALCULADO"] > 60000]
filtro7.to_csv('filtro7.csv')
filtro7
```

```python
#Filtro Y
filtro8 = df [(df["MARGEN_TOTAL_CALCULADO"] > 61000) & (df["FECHA_DOC"] == "2022-05-24")]
filtro8.to_csv('filtro8.csv')
filtro8
```

```python
#Filtro O
filtro9 = df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ") | (df["PRECIO_UNITARIO"] < 100)]
filtro9.to_csv('filtro9.csv')
filtro9
```

```python
#Filtro NOT
filtro10 = df[~(df["COSTO_TOTAL_CALCULADO"] < 5000)]
filtro10.to_csv('filtro10.csv')
filtro10
```
