```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leer el archivo de paises
df = pd.read_csv("listings.csv")
df
```

```python
#Filtro 1 los registros de 5 host name diferentes 
filtro1 = df.drop_duplicates('host_name').head()
filtro1.to_csv('filtro1.csv')
filtro1
```

```python
#Filtro 2 host que se unieron antes del 2019
filtro2 = df[(df["host_since"] <= '2019')]
filtro2.to_csv('filtro2.csv')
filtro2
```

```python
#Filtro 3 host que se tardan en contestar en 1 dia 
filtro3 = df[df["host_response_time"] == "within a day"]
filtro3.to_csv('filtro3.csv')
filtro3
```

```python
#filtro 4 registros que tengan al menos 5 accommodates
filtro4 = df[df["accommodates"] >= 5]
filtro4.to_csv('filtro4.csv')
filtro4
```

```python
#Filtro 5 registros de los tipos de cuarto 
filtro5 = df[(df["room_type"] != "Entire home") & (df["room_type"] != "Private room")]
filtro5.to_csv('filtro5.csv')
filtro5
```

```python
#Filtro 6 los registros que cuenten con 4 camas
filtro6 = df[df["beds"] <= 4]
filtro6.to_csv('filtro6.csv')
filtro6
```

```python
#Filtro 7 regitros con un baño o sean superhost, pero que cumpla la condicion
#de que todos sean instant_bookable
filtro7 = df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == 't') & (df["instant_bookable"] == 't'))]
filtro7.to_csv('filtro7.csv')
filtro7
```

```python
#Filtro 8 los registros donde reviews_scores_cleanliness sean menor a 3
# y review_scores_rating seran mayor de 3
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8.to_csv('filtro8.csv')
filtro8
```
